# Removing price outliers

```python
#Creating extra column for price_per_sqft
```

```python
df7=df6.copy()
df7
```

```python
df7['pp_sqft']=df7.price/df7.area_sqft
df7
```

```python
df7['pp_sqft'].describe()
```

Price_per_sqft can never be 0.000735 .so removing such kind of values

```python
def remove_pps_outliers(df):
    out_df=pd.DataFrame()
    
    for location,locational_df in df.groupby('location'):
        m=np.mean(locational_df['pp_sqft'])
        std=np.std(locational_df['pp_sqft'])
        reduced_df=locational_df[ (locational_df['pp_sqft']>(m-std))&
                                  (locational_df['pp_sqft']<(m+std)) ]
        out_df=pd.concat([out_df,reduced_df],ignore_index=True)
    return out_df
```

```python
df8 = remove_pps_outliers(df7)
df8
```

```python
df9=df8.drop('pp_sqft',axis='columns')
df9
```

# choosing best fit model using Grid search CV

```python
from sklearn import linear_model
```

```python
model_parameters={
     'Linear_regression':
     {
         "clf":LinearRegression(),'parameters':{'normalize':[False]}
     },
     'Linear_model.Lasso':
     {
         "clf":linear_model.Lasso(),"parameters":{'alpha':[1.0],"selection" : ['cyclic', 'random']}
     },
    'Linear_model.Ridge':
     {
         "clf":linear_model.Ridge(),"parameters":{'alpha':[1.0],
                                                  "solver" : ['auto', 'svd' ]}
     },
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
```

```python
cv=ShuffleSplit(n_splits=5,test_size=0.3,random_state=10)
```

```python
scores=[]
for model_name,mp in model_parameters.items():
    clf=GridSearchCV(mp['clf'],mp['parameters'],cv=cv,return_train_score=False)
    clf.fit(X_train,y_train)
    scores.append({
        'Model':model_name,'Best parametrs':clf.best_params_,'Best_score':clf.best_score_
    })
```

```python
scores_df=pd.DataFrame(scores)
scores_df
```

# Choosing LinearReession to be my final model

```python
def predict_price(sqft_area,bedrooms,location):
    index_loc=np.where(X.columns==location)[0][0]
    x=np.zeros(len(X.columns))
    x[0]=sqft_area
    x[1]=bedrooms
    if index_loc>=0:
        x[index_loc]=1
    return lr_model.predict([x])[0]
```
